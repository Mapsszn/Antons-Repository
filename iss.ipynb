```python
import requests
import matplotlib.pyplot as plt
import time

x_values = []
y_values = []

x_max = 180
y_max = 90
x_min = -180
y_min = -90


plt.title("ISS Position")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.plot(x_max, y_max)
plt.plot(x_min, y_min)

time_response = requests.get('https://api.wheretheiss.at/v1/satellites/25544')
time_data = time_response.json()
print(f"Start timestamp: " + str(time_data.get('timestamp')))

for i in range(0,100):
    response = requests.get('https://api.wheretheiss.at/v1/satellites/25544', timeout = 10)
    time.sleep(10)
    if response.status_code == 200:
        data = response.json()

        latitude = data.get('latitude')  

        longitude = data.get('longitude')  

        x_values.append(longitude)
        y_values.append(latitude)  

time_response_2 = requests.get('https://api.wheretheiss.at/v1/satellites/25544')
time2_data = time_response_2.json()
print(f"End timestamp: " + str(time2_data.get('timestamp')))

plt.plot(x_values, y_values)
plt.show()
```
